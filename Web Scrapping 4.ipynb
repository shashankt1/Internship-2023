```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome(r"C:\Users\shash\OneDrive\Desktop\chromedriver_win32\chromedriver.exe")
```

```python
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
rank =[]
song_name =[]
artist =[]
upload_date =[]
views =[]

for page in range(0,1):
    rank=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[1]')
    for i in rank:
        if 'text' in dir(i):
            rank.append(i.text)
            
            song_name= driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[2]')
            for j in song_name:
                if 'text' in dir(j):
                    song_name.append(j.text)
            artist=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[3]')
            for k in artist:
                if 'text' in dir(k):
                    artist.append(k.text)
            
            upload_date=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[5]')
            for l in upload_date:
                if 'text' in dir(l):
                    upload_date.append(l.text)
            
            views= driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[4]')
            for m in views:
                if 'text' in dir(m):
                    views.append(m.text)
```

```python
len(rank),len(song_name),len(upload_date),len(views),len(artist)
```

```python
df_youtbe =pd.DataFrame({"Rank": rank,"Name": song_name,"Artist":artist,"Upload date": upload_date,"Views":views})
```
